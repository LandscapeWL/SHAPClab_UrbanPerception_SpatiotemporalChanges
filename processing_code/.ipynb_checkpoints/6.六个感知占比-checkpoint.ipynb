```python
# 转换坐标
points = points.to_crs(epsg=3857)
```

```python
# unary_union 将一整列矢量合并为单独的一个对象
# convex_hull 返回每个几何对象的凸多边形
# centroid  返回每个几何对象的重心（几何中心）
cx,cy = points.unary_union.convex_hull.centroid.x,points.unary_union.convex_hull.centroid.y
```

```python
# 绘制所有点的中心
ax = points.plot()
ax.scatter(cx,cy,marker='*',color='red',s=200)
```

```python
# 根据几何中心绘制缓冲区
Point(cx,cy).buffer(300)
```

```python
# 绘制不同半径的中心
Point(cx,cy).buffer(600).difference(Point(cx,cy).buffer(300))
```

```python
def buffer():
    for i in range(0,13):
        bf = gpd.GeoDataFrame([[(i+1),Point(cx,cy).buffer((i+1)*1500).difference(Point(cx,cy).buffer(i*1500))]],
                              columns=['buffer','geometry'],crs='epsg:3857')
        yield bf
```

```python
# 将缓冲区进行合并
buffer = pd.concat(buffer())
```

```python
ax = points.plot(figsize=(10,10))
ax.scatter(cx,cy,marker='*',color='red',s=200)
buffer.plot(ax=ax,facecolor='none')
```

## 六种类型数据处理

```python
def get_score(perception):
    points = gpd.read_file(r'E:\202203PlacePulse\3.Zhangfan\data\5.预测街景感知分数\2015/XY2015_{}.shp'.format(perception))
    points = points.to_crs(epsg=3857)
    cx,cy = points.unary_union.convex_hull.centroid.x,points.unary_union.convex_hull.centroid.y

    def buffer():
        for i in range(0,13):
            bf = gpd.GeoDataFrame([[(i+1),Point(cx,cy).buffer((i+1)*1500).difference(Point(cx,cy).buffer(i*1500))]],
                                  columns=['buffer','geometry'],crs='epsg:3857')
            yield bf

    buffer = pd.concat(buffer())

    pointsx = gpd.sjoin(points,buffer,op='within')
    dt = pointsx.groupby(by='buffer').mean()['score']
    return dt
```

```python
perceptions = ['safer','beautiful','wealthier','livelier','depressing','boring']

df = pd.concat([get_score(perception) for perception in perceptions],axis=1)
```

```python
df.columns = perceptions
```

```python
df
```

```python
# df.sum(axis=1) 是对每一行求和
df.sum(axis=1)
```

# 标准化 计算每个感知分数在6个感知整体中的占比

```python
vals = df.values / df.sum(axis=1).values.reshape(13,1)
```

# 绘图代码（百分比）

```python
from matplotlib.pyplot import MultipleLocator
# 设置默认字体
plt.rcParams['font.sans-serif']=['Times New Roman'] 

# 字典results中储存绘图数据
results = {}
for i,j in enumerate(vals100):
    km1, km2 = 1.5*i ,1.5*(i+1)
    results['{}-{}km'.format(km1,km2)] = list(j)

# 设置图例的内容
category_names = perceptions

def survey(results, category_names):

    # 拿到标签
    labels = list(results.keys())
    # 拿到数据
    data = np.array(list(results.values()))
    # 逐步返回累积的和
    data_cum = data.cumsum(axis=1)
    
    # 设置色带 ['RdYlGn'] ['Spectral']
    # np.linespace后的是调整左右颜色深度
    category_colors = plt.colormaps['wl'](np.linspace(0, 1, data.shape[1]))

    # 设置绘图的尺寸
    fig, ax = plt.subplots(figsize=(10, 10),dpi=300,facecolor='white',edgecolor='white')   
    # 隐藏边框线
    ax=plt.gca()
    ax.spines['top'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['right'].set_visible(False)
    # 翻转坐标轴（y轴）
    ax.invert_yaxis()

    # 设置坐标轴范围
    ax.set_xlim(0, np.sum(data, axis=1).max())
    
    # 隐藏刻度
    #ax.axes.xaxis.set_ticks([])
    #ax.axes.yaxis.set_ticks([])
    
    # 设置xy轴的刻度值
    plt.xlabel("Human Perception (%)", fontsize=21)
    # 设置xy坐标轴的刻度
    plt.xticks(np.arange(0,120,25))
    # size设置刻度的长度
    plt.tick_params(labelsize=21,size=0)
    # 添加辅助线网格
    plt.grid(visible=True, axis='x', zorder=0)
    
    for i, (colname, color) in enumerate(zip(category_names, category_colors)):
        
        widths = data[:, i]
        starts = data_cum[:, i] - widths
        
        # y:y轴序列   width:条的宽度  left:杆左边的x坐标  height:条的高度 
        # label:设置标签  locor:设置颜色  edgecolor:设置柱状图边框色
        rects = ax.barh(y=labels, width=widths, left=starts, height=0.45,
                        label=colname, color=color, edgecolor='black',zorder=100)

        # 控制柱状图图例显示
        #r, g, b, _ = color
        #text_color = 'white' if r * g * b < 0.5 else 'darkgrey'
        #ax.bar_label(rects, label_type='center', color=text_color)
    
    # 设置图例 ncol:图例一行放几个
    # bbox_to_anchor:指loc的中点位置
    ax.legend(ncol=3, bbox_to_anchor=(0.5, 1.05),loc='center', fontsize=20,edgecolor='white')
 
    
    return fig, ax


survey(results, category_names)
plt.show()
```
